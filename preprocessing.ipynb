<center><h2>SENTIMENT ANALYSIS</h2></center>


1. The code performs sentiment analysis on a given tweet.
2. The tweet is preprocessed by replacing mentions with "@user" and URLs with "http".
3. A pre-trained sentiment analysis model, "cardiffnlp/twitter-roberta-base-sentiment", is loaded along with its tokenizer.
4. The model is used to predict the sentiment of the preprocessed tweet, with scores for negative, neutral, and positive sentiment.
5. The scores are passed through a softmax function to convert them to probabilities.
6. The results are printed and plotted as a bar graph, showing the probability of each sentiment category.


<h3>Inputs:</h3>

    CSV file located at 'data/sample_input.csv', containing tweets to be analyzed
    Pretrained sentiment analysis model 'cardiffnlp/twitter-roberta-base-sentiment'
    Labels for sentiment categories: 'Negative', 'Neutral', 'Positive'

<h3>Output:</h3>

    Sentiment analysis results for each tweet in the CSV file, with a breakdown of the scores for each sentiment category
    A bar plot showing the sentiment analysis results for each tweet

```python
import csv
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax
import matplotlib.pyplot as plt

tweets = []

# read CSV file and extract tweets
with open('data/sample_input.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        tweet = row[3] # assuming the tweet is encoded in utf-8
        tweets.append(tweet)

# load model and tokenizer
roberta = "cardiffnlp/twitter-roberta-base-sentiment"

model = AutoModelForSequenceClassification.from_pretrained(roberta)
tokenizer = AutoTokenizer.from_pretrained(roberta)

labels = ['Negative', 'Neutral', 'Positive']

# analyze each tweet and print results
for tweet in tweets:
    # preprocess tweet
    tweet_words = []
    for word in tweet.split(' '):
        if word.startswith('@') and len(word) > 1:
            word = '@user'
        elif word.startswith('http'):
            word = "http"
        tweet_words.append(word)
    tweet_proc = " ".join(tweet_words)

    # sentiment analysis
    encoded_tweet = tokenizer(tweet_proc, return_tensors='pt')
    output = model(**encoded_tweet)

    scores = output[0][0].detach().numpy()
    scores = softmax(scores)

    # print results
    print(f"Tweet: {tweet}")
    for i in range(len(scores)):
        l = labels[i]
        s = scores[i]
        print(f"{l}: {s}")
    print()

    # plot results
    fig, ax = plt.subplots()
    ax.bar(labels, scores)
    ax.set_title('Sentiment Analysis Results')
    ax.set_xlabel('Sentiment')
    ax.set_ylabel('Score')
    plt.show()
```

<p>
    <h3>Defining a function to check for hypernyms: </h3>
    
    The code defines a function called check_for_hypernim() which takes a token as input and returns its hypernyms. The function uses the WordNet corpus from nltk.corpus to find the hypernyms of the token.
    

</p>

```python
import csv
stop_words = set(stopwords.words('english'))


def check_for_hypernim(token):
    hypernims = []
    for i in range(15):
        try:
            hypernims1 = []
            for i, j in enumerate(wn.synsets(token)):
                for l in j.hypernyms():
                    hypernims1.append(l.lemma_names()[0])
            token = hypernims1[0]
            hypernims.append(hypernims1)
        except IndexError:
            hypernims.append([token])

    return hypernims

```

<h3>Filtering out stop words: </h3>
    
    The code filters out the stop words from the extracted words.
    
<h3>Part of speech tagging: </h3>
    
    The code performs part-of-speech (POS) tagging on the filtered words using nltk.pos_tag() function.
    
<h3>Filtering out unnecessary parts of speech: </h3>
    
    The code filters out certain parts of speech that are unlikely to have hypernyms, such as determiners, conjunctions, interjections, and prepositions.
<h3>
Writing hypernyms to a CSV file: </h3>


    The code writes the hypernyms of each word to a new CSV file called "sample_hypernyms.csv". It uses the check_for_hypernim() function to find the hypernyms of each word and writes them to the CSV file.

<h3>Inputs:</h3>


1. "data/sample_input.csv": A CSV file containing text data.
    
2. English stop words: A set of predefined stop words in the English language.
</p>

```python
file1 = open("data/sample_input.csv",encoding='latin1')
line = file1.read()
words = line.split()
with open("results/sample_hypernyms.csv", 'w',encoding='latin1') as csv_file:
    filewriter = csv.writer(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL, lineterminator='\n')
    filtered_sentence = []

    x = []
    for r in words:

        if not r in stop_words:
            filtered_sentence.append(r)
    tagged = nltk.pos_tag(filtered_sentence)
    for i in tagged:
        if len(i[0]) != 0 or len(i[0]) != 1:
            if i[1] == 'IN' or i[1] == 'DT' or i[1] == 'CD' or i[1] == 'CC' or i[1] == 'EX' or i[1] == 'MD' or   i[1] == 'WDT' or i[1] == 'WP' or i[1] == 'UH' or i[1] == 'TO' or i[1] == 'RP' or i[1] == 'PDT' or i[1] == 'PRP' or i[1] == 'PRP$' or i[0] == 'co':
                # print(i[0])
                continue
            else:
                x.append(i[0].rstrip(".,?!"))
    for i in x:
        l = []
        l.append(i)
        hype = check_for_hypernim(i)
        if len(hype) == 0:
            hype.append(i)  # 1
            hype.append(i)  # 2
            hype.append(i)  # 3
            hype.append(i)  # 4
            hype.append(i)  # 5
            hype.append(i)  # 6
            hype.append(i)  # 7
            hype.append(i)  # 8
            hype.append(i)  # 9
            hype.append(i)  # 10
            hype.append(i)  # 11
            hype.append(i)  # 12
            hype.append(i)  # 13
            hype.append(i)  # 14
            hype.append(i)  # 15
        for hyper in hype:
            l.append(hyper[0])
        filewriter.writerow(l)
    csv_file.close()
```
